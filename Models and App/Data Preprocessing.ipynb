```python
x=df.drop('Heart Attack Risk',axis=1)
y=df['Heart Attack Risk']
```

```python
from sklearn.preprocessing import LabelEncoder

# Identify categorical columns
categorical_cols = x.select_dtypes(include=['object']).columns

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Dictionary to store mappings for each categorical column
column_mappings = {}

# Converting categorical columns to numeric values using Label Encoding
for col in categorical_cols:
    # Handling NaN values: Optionally, you could replace them or handle them
    x[col] = x[col].fillna('Missing')  # Replacing NaN with a placeholder if needed
    
    # Fit and transform the data
    # Fit the encoder to the column's unique values
    label_encoder.fit(x[col])

    # Transform the data using the fitted encoder
    x[col] = label_encoder.transform(x[col])

    
    # Save the mapping for each column
    column_mappings[col] = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

# Print the data types of columns after encoding
print(x.dtypes)
```

```python
categorical_cols
```

```python
# Print the mappings for each categorical column
for col, mapping in column_mappings.items():
    print(f"Mapping for column '{col}':\n", mapping)
```

```python
# Save the encoder for later use
import joblib
joblib.dump(label_encoder, 'label_encoder.pkl')
```

```python
# Outlier visualization
for col in x.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x[col], color='skyblue')
    plt.title(f"{col} Before Outlier Treatment")
    plt.show()
```

```python
x['Sedentary Hours Per Day'].describe()
```

```python
# Outlier handling using IQR
for col in x.columns:
    Q1, Q3 = np.percentile(x['Sedentary Hours Per Day'], [25, 75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    x['Sedentary Hours Per Day'] = np.where(x['Sedentary Hours Per Day'] > upper, upper,\
                                               np.where(x['Sedentary Hours Per Day'] < lower, lower, x['Sedentary Hours Per Day']))
```

```python
# Outlier visualization
plt.figure(figsize=(8, 5))
sns.boxplot(x['Sedentary Hours Per Day'], color='skyblue')
plt.title(f"Sedentary Hours Per Day After Outlier Treatment")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
sns.countplot(x=y_train)
plt.title("Class Distribution in Training Set")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler

# Scale your data (this will help with logistic regression convergence)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize Logistic Regression Classifier
logreg_clf = LogisticRegression(max_iter=500, random_state=42)

# Fit the model with scaled data
logreg_clf.fit(X_train_scaled, y_train)

# Make Predictions
y_pred = logreg_clf.predict(X_test_scaled)

# Evaluating the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
#Create and train the random forest classifier
rf_clf=RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
rf_clf.fit(X_train, y_train)

#Make Predictions
y_pred=rf_clf.predict(X_test)

#Evaluating the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n",confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
xgb_clf = XGBClassifier(eval_metric='logloss', random_state=42)
xgb_clf.fit(X_train, y_train)

#Make Predictions
y_pred=xgb_clf.predict(X_test)

#Evaluating the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n",confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import VotingClassifier

# Define the Voting Classifier
voting_clf = VotingClassifier(
    estimators=[('rf', rf_clf), ('logreg', logreg_clf), ('xgb', xgb_clf)],
    voting='soft'  # Use soft voting to take weighted average of probabilities
)

# Fit the model
voting_clf.fit(X_train_scaled, y_train)

# Make predictions
y_pred = voting_clf.predict(X_test_scaled)

# Evaluate model performance
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n",confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Extract and display feature importances
importances = xgb_clf.feature_importances_
features = x.columns  # Assuming 'X' is your feature set

print("\nFeature Importances:")
for feature, importance in zip(features, importances):
    print(f"{feature}: {importance:.4f}")
```
